```python
import os
from gensim import corpora
from gensim.models import LsiModel
import nltk
nltk.download('stopwords')
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from gensim.models.coherencemodel import CoherenceModel
import matplotlib.pyplot as plt
```

```python
documents = []
doc_list = os.listdir("./NL/")
len(doc_list)

for document in doc_list:
    with open (f"./NL/{document}") as f:
        documents.append(f.read())
len(documents)
```

```python
tokenizer = RegexpTokenizer(r'\w+')

nl_stopwords = set(stopwords.words("dutch"))
len(nl_stopwords)
print(nl_stopwords)
stemmer = nltk.stem.SnowballStemmer("dutch")
```

```python
for i in range(len(documents)):
    documents[i] = documents[i].lower()
```

```python
tokenized = []
for document in documents:
    tokenized.append(tokenizer.tokenize(document))
len(tokenized)
```

```python
for i in range(len(tokenized)):
    print(f"before{len(tokenized[i])}")
    for word in tokenized[i]:
        
        if word in nl_stopwords:
            tokenized[i].remove(word)
    print(f"after{len(tokenized[i])}")
```

```python
stemmed = []
for document in documents:
    stemmed.append(stemmer.stem(document))
len(stemmed)
```

```python
import transformers
bert_tokenizer = transformers.AutoTokenizer.from_pretrained("yhavinga/t5-v1.1-base-dutch-cased")
model = transformers.AutoModelWithLMHead.from_pretrained("yhavinga/t5-v1.1-base-dutch-cased")



```

```python
documents[0]
```

```python
inputs = bert_tokenizer.encode(f"summarize: {documents[0]}",
return_tensors='pt',
max_length=512,
truncation=True)
```

```python
summary_ids = model.generate(inputs, max_length=383, min_length=80, length_penalty=5., num_beams=2)
```

```python
with open ("bert_test.txt","w+",encoding="utf-8") as f:
    print(bert_tokenizer.decode(summary_ids[0],skip_special_tokens=True),file=f)
```

```python
dictionary = corpora.Dictionary([stemmed])
doc_term_matrix = dictionary.doc2bow(stemmed)
```
