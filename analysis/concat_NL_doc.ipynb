```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
#load data 
df_nl_extra_fixed = pd.read_csv("NL_doc_paragraphs_extra_fixed.csv")
df_nl_extra_mean_fixed = pd.read_csv("NL_doc_paragraphs_extra_mean_fixed.csv")
df_nl_extra_mean = pd.read_csv("NL_doc_paragraphs_extra_mean.csv")
df_nl_extra = pd.read_csv("NL_doc_paragraphs_extra.csv")
df_nl_para = pd.read_csv("NL_doc_paragraphs.csv")
df_new_processed = pd.read_csv("new_processed_data.csv")
```

```python
# insert new column 'sub_part' to data set df_nl_para
df_nl_para.insert(2,'sub_part','',True)
```

```python
# insert new column 'sub_part' to data set df_nl_extra
df_nl_extra.insert(2,'sub_part','',True)
```

```python
df_nl_extra_fixed.reset_index(drop=True, inplace=True)
df_nl_extra_mean_fixed.reset_index(drop=True, inplace=True)
df_nl_extra_mean.reset_index(drop=True, inplace=True)
df_nl_extra.reset_index(drop=True, inplace=True)
df_nl_para.reset_index(drop=True, inplace=True)
df_new_processed.reset_index(drop=True, inplace=True)
```

```python
# concate csv files 
df_concate = pd.concat(map(pd.read_csv, ["NL_doc_paragraphs_extra_fixed.csv","new_processed_data.csv","NL_doc_paragraphs_extra_mean.csv","NL_doc_paragraphs_extra_mean_fixed.csv","NL_doc_paragraphs.csv", "NL_doc_paragraphs_extra.csv"]))
```

```python
df_concate.head()
df_concate.drop("Unnamed: 0",axis=1,inplace=True)
df_concate
```

```python
# save file a csv
df_concate.to_csv("NL_final_doc_paragraphs.csv", index = False, header=True)
```
